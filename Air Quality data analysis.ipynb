```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# -------------------------------
# Load the datasets
# -------------------------------
city_day = pd.read_csv('city_day.csv')
city_hour = pd.read_csv('city_hour.csv')

# -------------------------------
# Basic Pandas operations
# -------------------------------
print("City Day Data Head:\n", city_day.head())
print("\nCity Hour Columns:\n", city_hour.columns)

# Convert date columns to datetime
city_day['Date'] = pd.to_datetime(city_day['Date'])
city_hour['Datetime'] = pd.to_datetime(city_hour['Datetime'])

# Filter for a specific city
city = 'Delhi'
city_day_filtered = city_day[city_day['City'] == city]
city_hour_filtered = city_hour[city_hour['City'] == city]

# -------------------------------
# Handle missing data
# -------------------------------
city_day_filtered.fillna(method='ffill', inplace=True)
city_hour_filtered.fillna(method='bfill', inplace=True)

# -------------------------------
# NumPy operations on AQI
# -------------------------------
aqi_array = city_day_filtered['AQI'].dropna().values
print("\nNumPy Summary Statistics on AQI for", city)
print("Mean AQI:", np.mean(aqi_array))
print("Max AQI:", np.max(aqi_array))
print("Min AQI:", np.min(aqi_array))

# -------------------------------
# Matplotlib Visualization
# -------------------------------
# AQI over time
plt.figure(figsize=(12, 5))
plt.plot(city_day_filtered['Date'], city_day_filtered['AQI'], color='purple', label='AQI')
plt.xlabel('Date')
plt.ylabel('AQI')
plt.title(f'AQI Trend in {city} (Daily)')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

# PM2.5 vs PM10 scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(city_hour_filtered['PM2.5'], city_hour_filtered['PM10'], alpha=0.5, color='orange')
plt.xlabel('PM2.5')
plt.ylabel('PM10')
plt.title(f'PM2.5 vs PM10 in {city} (Hourly)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the datasets
city_hour = pd.read_csv('city_hour.csv')
city_day = pd.read_csv('city_day.csv')

# Clean and preprocess the data
city_hour['Datetime'] = pd.to_datetime(city_hour['Datetime'])
city_day['Date'] = pd.to_datetime(city_day['Date'])

# Drop columns with >50% missing values
city_hour = city_hour.dropna(thresh=len(city_hour) * 0.5, axis=1)
city_day = city_day.dropna(thresh=len(city_day) * 0.5, axis=1)

# Fill remaining missing values
city_hour.fillna(method='ffill', inplace=True)
city_hour.fillna(method='bfill', inplace=True)
city_day.fillna(method='ffill', inplace=True)
city_day.fillna(method='bfill', inplace=True)

# Visualizations for a selected city
sample_city = 'Delhi'
city_hour_sample = city_hour[city_hour['City'] == sample_city]
city_day_sample = city_day[city_day['City'] == sample_city]

# 1. Line plot of PM2.5 and PM10 over time
plt.figure(figsize=(14, 6))
plt.plot(city_hour_sample['Datetime'], city_hour_sample['PM2.5'], label='PM2.5')
plt.plot(city_hour_sample['Datetime'], city_hour_sample['PM10'], label='PM10')
plt.title(f'Hourly PM2.5 & PM10 Levels in {sample_city}')
plt.xlabel('Datetime')
plt.ylabel('Concentration (g/m)')
plt.legend()
plt.tight_layout()
plt.show()

# 2. Boxplot of AQI by AQI Bucket
plt.figure(figsize=(10, 6))
sns.boxplot(data=city_day_sample, x='AQI_Bucket', y='AQI', palette='coolwarm')
plt.title(f'AQI Distribution by Category in {sample_city}')
plt.xlabel('AQI Category')
plt.ylabel('AQI Value')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 3. Heatmap of correlations
plt.figure(figsize=(12, 8))
corr = city_day_sample.drop(columns=['City', 'Date', 'AQI_Bucket']).corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title(f'Correlation Heatmap of Pollutants in {sample_city}')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Load the dataset
city_day = pd.read_csv('city_day.csv')

# Parse date column
city_day['Date'] = pd.to_datetime(city_day['Date'])

# Drop columns with more than 50% missing values
city_day = city_day.dropna(thresh=len(city_day) * 0.5, axis=1)

# Fill missing values
city_day.fillna(method='ffill', inplace=True)
city_day.fillna(method='bfill', inplace=True)

# Filter data for a sample city
sample_city = 'Delhi'
df_city = city_day[city_day['City'] == sample_city].copy()

# Drop non-numeric columns for PCA
features = df_city.drop(columns=['City', 'Date', 'AQI_Bucket'])

# Standardize the features
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

# Apply PCA
pca = PCA(n_components=2)
principal_components = pca.fit_transform(scaled_features)

# Create a DataFrame with the principal components
pca_df = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2'])
pca_df['AQI_Bucket'] = df_city['AQI_Bucket'].values

# Plot the PCA result
plt.figure(figsize=(10, 6))
sns.scatterplot(data=pca_df, x='PC1', y='PC2', hue='AQI_Bucket', palette='Set2', alpha=0.7)
plt.title(f'PCA of Air Quality Features in {sample_city}')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(title='AQI Category', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
import numpy as np

# ----------- Scalars -----------
# A scalar is a single number
scalar_a = 42
scalar_b = 3.14

print("Scalar a:", scalar_a)
print("Scalar b:", scalar_b)

# ----------- Vectors -----------
# A vector is an array (1D) of numbers
vector_pm25 = np.array([50, 80, 60, 45, 72])  # e.g., PM2.5 levels over 5 days
vector_pm10 = np.array([100, 120, 90, 110, 95])  # PM10 over same 5 days

print("\nVector PM2.5:", vector_pm25)
print("Vector PM10:", vector_pm10)

# Vector operations (element-wise)
sum_vector = vector_pm25 + vector_pm10
diff_vector = vector_pm10 - vector_pm25
scaled_vector = 0.5 * vector_pm25

print("\nSum of vectors (PM2.5 + PM10):", sum_vector)
print("Difference (PM10 - PM2.5):", diff_vector)
print("Scaled PM2.5 (0.5):", scaled_vector)

# Dot product of vectors
dot_product = np.dot(vector_pm25, vector_pm10)
print("Dot Product (PM2.5  PM10):", dot_product)

# Magnitude (norm) of a vector
magnitude_pm25 = np.linalg.norm(vector_pm25)
print("Magnitude (norm) of PM2.5 vector:", magnitude_pm25)

# ----------- Visualization (Optional) -----------
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.plot(vector_pm25, label='PM2.5', marker='o')
plt.plot(vector_pm10, label='PM10', marker='s')
plt.title("PM2.5 and PM10 as Vectors over Days")
plt.xlabel("Day Index")
plt.ylabel("Concentration")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# ----------- 1. Undirected Graph -----------
G_undirected = nx.Graph()

# Add edges (nodes will be auto-added)
G_undirected.add_edges_from([
    ('Delhi', 'Mumbai'),
    ('Delhi', 'Kolkata'),
    ('Mumbai', 'Chennai'),
    ('Kolkata', 'Chennai')
])

plt.figure(figsize=(6, 4))
nx.draw(G_undirected, with_labels=True, node_color='lightblue', edge_color='gray', node_size=1500)
plt.title("Undirected Graph")
plt.show()

# ----------- 2. Directed Graph -----------
G_directed = nx.DiGraph()

G_directed.add_edges_from([
    ('Delhi', 'Mumbai'),
    ('Mumbai', 'Kolkata'),
    ('Kolkata', 'Chennai'),
    ('Chennai', 'Delhi')
])

plt.figure(figsize=(6, 4))
nx.draw(G_directed, with_labels=True, node_color='lightgreen', edge_color='black', node_size=1500, arrows=True)
plt.title("Directed Graph")
plt.show()

# ----------- 3. Weighted Graph -----------
G_weighted = nx.Graph()

# Add edges with weights
G_weighted.add_weighted_edges_from([
    ('Delhi', 'Mumbai', 8),
    ('Delhi', 'Kolkata', 5),
    ('Mumbai', 'Chennai', 10),
    ('Kolkata', 'Chennai', 7)
])

plt.figure(figsize=(6, 4))
pos = nx.spring_layout(G_weighted)
nx.draw(G_weighted, pos, with_labels=True, node_color='lightcoral', edge_color='gray', node_size=1500)
labels = nx.get_edge_attributes(G_weighted, 'weight')
nx.draw_networkx_edge_labels(G_weighted, pos, edge_labels=labels)
plt.title("Weighted Undirected Graph")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load and clean the data
city_day = pd.read_csv('city_day.csv')
city_day['Date'] = pd.to_datetime(city_day['Date'])

# Drop columns with >50% missing values
city_day = city_day.dropna(thresh=len(city_day) * 0.5, axis=1)

# Fill remaining missing values
city_day.fillna(method='ffill', inplace=True)
city_day.fillna(method='bfill', inplace=True)

# Select a sample city
sample_city = 'Delhi'
city_ts = city_day[city_day['City'] == sample_city].copy()

# Set the date as index for time series plotting
city_ts.set_index('Date', inplace=True)

# Plot PM2.5 over time
plt.figure(figsize=(14, 6))
plt.plot(city_ts.index, city_ts['PM2.5'], label='PM2.5', color='orange')
plt.title(f'Daily PM2.5 Trend in {sample_city}')
plt.xlabel('Date')
plt.ylabel('PM2.5 (g/m)')
plt.grid(True)
plt.tight_layout()
plt.show()

# Plot AQI over time
plt.figure(figsize=(14, 6))
plt.plot(city_ts.index, city_ts['AQI'], label='AQI', color='green')
plt.title(f'Daily AQI Trend in {sample_city}')
plt.xlabel('Date')
plt.ylabel('AQI Value')
plt.grid(True)
plt.tight_layout()
plt.show()

# Optional: Rolling average for smoother trend
city_ts['PM2.5_Rolling'] = city_ts['PM2.5'].rolling(window=30).mean()
city_ts['AQI_Rolling'] = city_ts['AQI'].rolling(window=30).mean()

# Plot rolling averages
plt.figure(figsize=(14, 6))
plt.plot(city_ts.index, city_ts['PM2.5_Rolling'], label='PM2.5 (30-day Avg)', color='red')
plt.plot(city_ts.index, city_ts['AQI_Rolling'], label='AQI (30-day Avg)', color='blue')
plt.title(f'{sample_city} - 30-Day Rolling Average of PM2.5 and AQI')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
